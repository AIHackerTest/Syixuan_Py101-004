**习题 5：更多变量和打印**

格式化字符 %r：不管什么都打印出来
其他格式化字符：
%s 字符串 (采用str()的显示)；%r 字符串 (采用repr()的显示)；%c 单个字符；%b 二进制整数；%d 十进制整数；%i 十进制整数；%o 八进制整数；%x 十六进制整数；%e 指数 (基底写为e)；%E 指数 (基底写为E)；%f 浮点数；%F 浮点数，与上相同；%g 指数(e)或浮点数 (根据显示长度)；%G 指数(E)或浮点数 (根据显示长度)%% 字符"%"

**习题 6：字符串和文本**

字符串：用 “ ”或 ‘’，括起来的部分
字符串可以包含格式化字符 %s 等
字符串可以执行连接运算 —— 如 w + e

%r 用来 debug, 又称 representation，因为它会显示变量的原始数据
%s 及其他字符串用来向用户显示输出

**习题 9：打印，打印，打印**

"\n" 用来换行
""" """ 之间可输入多行

**习题 10：转义字符**

反斜杆 \ backslash

转义序列
\\ 反斜杆 ；\' 单引号；\" 双引号；\a 响铃符
\b 退格符；\f 进纸符；\n 换行符；\N{name} Unicode 数据库中的字符名
\r ASCII 回车符；\t ASCII 水平制表符；\uxxxx 值为 16 位十六进制值 xxxx 的字符
\Uxxxxxxxx 值为 32 位十六进制值 xxxx 的字符； \v 垂直制表符
\ooo 值为八进制值 ooo 的字符；\xhh 值为十六进制数 hh 的字符

**习题 11：提问**

x = int(raw_input())
input() 是函数，会把你输入的东西当做 Python 的代码来处理；而 raw_input() 只是输入一个值
powershell 下输入 python -m pydoc raw_input 可查看其功能用法

**习题 13：参数、解包、变量**

把 argv 中的东西解包，将所有参数依次赋予左边的变量名
在 powershell 中输入参数，第一个参数永远是文件脚本名
import 进来的如 sys 称模组（modules），也称库（libraries）
参数在用户执行命令时输入用 argv；在脚本运行过程中输入用 raw_input()

**习题 15：读取文件**

print txt.read() 读取文件 
txt = open(filename) 返回的不是文件内容，而是一个叫 "file object" 的东西，它像一个磁带机或 DVD 机。你可以随意访问内容的任何位置，并读取这些内容，但这个 object 不是内容本身

**习题 16：读写文件**

- close——关闭文件，类似于保存
- read——读取文件内容，把结果赋给一个变量
- readline——读取文本文件中的一行
- truncate——清空文件
- write(stuff)——将 stuff 写入文件

**习题 18：命名、变量、代码、函数**

函数做的三件事
- 给代码片段命名，就像“变量”给字符串和数字命名一样
- 可以接受参数，就像脚本接受 argv 一样
- 通过使用 #1 和 #2，可以创建“微型脚本”和“小命令”

*args：把函数所有的参数都接受进来，然后放到名字叫 args 的列表中去；和 argv 类似
exit()：可以中断某个程序，而其中的数字参数则用来表示程序是否是碰到错误而中断。
exit(1)表示发生了错误，而 exit(0)表示程序正常退出
